```python
import time

tweet_id = "1546207654699204611"

base_url = "https://api.twitter.com/2/tweets"
endpoint = f"/{tweet_id}/retweeted_by"

token = "YOUR_TOKEN_HERE"
headers = {"Authorization": f"Bearer {token}"}

res = requests.get(base_url + endpoint, headers=headers)

next_token = res.json()["meta"].get("next_token")
data = res.json()["data"]
retweets_id = list(map(lambda x: x["id"], data))


while next_token is not None:
    res = requests.get(
        base_url + endpoint,
        headers=headers,
        params={"max_results": 100, "pagination_token": next_token},
    )

    if res.json().get("data") is None:
        break
    retweets_id.extend(list(map(lambda x: x["id"], res.json()["data"])))
    next_token = res.json()["meta"].get("next_token")

len(retweets_id)
```
